## csv 모듈을 이용한 쓰기 방법

```python
import csv
f=open('test.csv','w') # 새롭게 행이 추가될때마다 빈행이 생김
data=csv.writer(f)
data.writerow(['첫번째','두번째','세번째'])
data.writerow([1,2,3])
f.close()
```

```python
f=open('test1.csv','w', newline='') #newline=''로 빈행 없애기
data=csv.writer(f)
data.writerow(['첫번째','두번째','세번째'])
data.writerow([1,2,3])
f.close()
```
