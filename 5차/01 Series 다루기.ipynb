###  * Series data 
> 1. pd.Series([1,2,3,4]) <br>
> 2. pd.Series([1,2,3,4], index=['a','b','c','d']

```python
# Series data 
a=pd.Series([1,2,3,4,])
print(a)
b=pd.Series([1,2,3,4], index=['a','b','c','d'])
print(b)
```

```python
#type() 
print(type(a))
print(type(b))
```

### * Series  
 > Series  index values .

```python
print(b.index)
print(b.values)
print(a.index)
print(a.values)
```

### *   
> Series index index 

```python
#
print(b['b'])
print(b[1])

#  2 
print(b[['a','c']])
print(b[[1,3]])

#
print(b[0:2])
print(b[:-1])
print(b[-2:])
print(b['a':'c'])

#  
b['a':'c']=100
print(b)
```
