#### Age 변수 변환하기
 - 변환 로직을 함수로 만든 후, apply 함수로 적용

```python
import math
```

```python
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data
```

```python
train_data['Age'].apply(age_categorize)
```
