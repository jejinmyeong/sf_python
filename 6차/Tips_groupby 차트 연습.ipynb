```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
tips = sns.load_dataset('tips')
```

```python
display(tips.describe())
display(tips.shape)
display(tips.head())  #주피터노트북에서 print 대신에 display로 해도됨
print(tips.info())
```

```python
#요일별 팁 합을 구하고 차트로 나타내기
tips_sum_by_day = tips.groupby('day').tip.sum()
tips_sum_by_day
```

```python
tips_sum_by_day.sort_values().plot(kind='bar')
```

```python
tips_sum_by_day.plot(kind='pie',autopct='%1.1f%%')
```

```python
#요일별, 성별 팁 합을 구하고 차트로 나타내기
a=tips.groupby(['day','sex'])[['tip']].sum()
```

```python
a.unstack(1)
```

```python
a.unstack(1)['tip'].plot(kind='bar',stacked=True)
```

```python
#흡연 유무에 Tip집계
tips.groupby('smoker').tip.sum().plot(kind='bar')
```
