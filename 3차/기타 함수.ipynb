```python
# map 함수 활용
def test(value):
    return value*2

list(map(test,[3,4,5,6]))
```

```python
# lambda함수
# [형식] lamda 매개변수들:식

# [방법1] 함수 지정
plus=lambda x:x+10
plus(100)

# [방법2] (lambda 매개변수들:식)(인수들)
(lambda x:x+10)(1)
```

```python
# map 함수와 lambda 함수
list(map(lambda x: x+10, [1,2,3,4]))
```

```python
#zip 함수(1)
a=[1,2,3,4]
b=['a','b','c','d']
print(list(zip(a,b)))
print(dict(zip(a,b)))
```
