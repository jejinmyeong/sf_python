```python
#  (find  index)
#rfind  rindex :    
s='apple pineapple'
print(s.find('pine'))
print(s.find('a',5)) # s[5:]  
print(s.find('ap',5,-1)) #s[5:-1] 
print(s.find('5')) #   -1 
print(s.index('5')) #index find      
```

```python
#   (count)
s='apple pineapple'
s.count('a')
```
