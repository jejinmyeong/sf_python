# 데이터 타입, 변수, 그리고 출력

## 1. 주요 데이터 타입

- 크게 문자, 숫자, 참/거짓(boolean) 으로 나뉩니다.
- 문자는 '문자' 또는 "문자" 와 같이 따옴표와 함께 써야 합니다. 
- 숫자는 숫자 그대로 쓰면 됩니다.
- 참/거짓(boolean)은 True/False 로 씁니다. (대소문자 구분합니다.)

### 1) 숫자형 자료


>
> - 정수(int)
>    - (10진수) 123, -234, 0 <br>
>    - (2진수_binary) 0b10, (8진수_Octal) 0o10, (16진수_Hexadecimal) 0x10 <br>
> - 실수(float): 123.45, -1234.5,3.4e10 <br>
> - 복소수(complex): x=3-4j

### 산술 연산자
> \+ (더하기),   -(빼기),   \*(곱하기),   /(나누기),   //(정수 몫),   %(나머지),   \**(거듭제곱)


### 2) 문자열 자료형




#### 문자 데이터 연산자 : + (문자열 연결), * (문자열 반복) <br>
> 예: 'abcd'+'가나다' -> 'abcd가나다'<br>
> 예: '=' * 10  -> '=========='<br>
> 예: 'abc' + 3 ->에러,  'abc'+str(3) =->'abc3'

## 2. 변수와 데이터 타입
* 변수(variable): 값을 저장하는 공간(메모리)의 위치를 가리키는 이름표
### 변수명 = 값
 > - a=1<br>
 > - a=1;b=2;c=3<br>
 > - a=b=3 <br>
 > - a,b=3,4<br>
 > <br>
 > 변수 재할당하면 가장 마지막에 할당된 값을 기억함<br>
 > <br>
 > - 변수 삭제 : del 변수명

### < 변수명 작성 규칙 >
* 변수이름은 영문자, 숫자, \_로 구성
* 숫자로 시작랄 수 없다.
* 파이썬 키워드(예약어) 로 변수명으로 사용할 수 없다.<br>
 예) and, del, from, if.....<br>

<br>
* 잘못된 변수명: A-3, A 3, 3A, A.3, if <p>
<p>
<b>[변수의 이름] </b>
* 의미있는 이름을 사용하는 것이 좋다.
* 소문자와 대문자를 구분한다.
* 낙타표기법(camelcase): 첫단어는 소문자로 표기하고, 새로운 단어 첫음절만 대문자로 표기하는 방법<br> 
    (예: myName, manAge)
* 뱀표기법(snakecase): 언더바로 구분 (예:my_name, man_age)<br>


## 3. 데이터 타입확인: type()
> type() 라는 함수를 통해, 데이터 타입 확인이 가능합니다.<br>
> 데이터 타입에 따라, 변수를 다루는 방법이 달라지므로, 꼭 데이터 타입을 확인하세요

- str: 문자
- int: 정수 (숫자는 정수 또는 부동소숫점 으로 세분화 됩니다.)
- float: 부동소숫점
- bool: 참/거짓(boolean)

## input()함수로 키보드 입력값을 변수에 할당하기

- 변수=input()
